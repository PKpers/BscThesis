# Let's compare the two methods of sampling

## Method 1: Normal distribution 
Generate 3 numbers $x_{1},x_{2}, x_{3}$ normaly distributed. Then the end point of vector 
$$
\vec{x} = 
\frac{1}{\sqrt{x_{1}^{2} + x_{2}^{2} + x_{3}{2}}} 
\begin{pmatrix}
x_{1} \\
x_{2} \\
x_{3} \\
\end{pmatrix}
$$
Will be normally distributed in the surfafce of the unitary circle unitary sphere

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import display_markdown, Markdown
from matplotlib import pyplot as plt

def make_nparray(_list_):
    _list_ = np.array(_list_)
    return _list_
```

```python
points = np.random.normal(0,1,(500,3))
x, y, z = ([], [], [])
for point in points:
    xi,yi,zi = point
    x.append(xi)
    y.append(yi)
    z.append(zi)
#
xyz = (x,y,z)
xyz_np = []
for i in xyz:
    inp = make_nparray(i)
    xyz_np.append(inp)
#
x, y, z = xyz_np
display(Markdown("### Plot the distributions of normally distributed x, y, z"))
# Lets see how the numbers are distributed
fig_hist, axs = plt.subplots(
    1, 3,sharey=True, tight_layout=True,  figsize=(17,7)
)

n_bins = 50
axs[0].hist(x, bins=n_bins,histtype='step')
axs[0].set_title('x distribution ', fontsize=20)
axs[1].hist(y, bins=n_bins, histtype='step')
axs[1].set_title('y distribution ', fontsize = 20)
axs[2].hist(z, bins=n_bins,  histtype='step')
axs[2].set_title('z distribution ', fontsize=20)
axs[2].set_xlabel('data',fontsize=20, loc='right')
axs[0].set_ylabel('counts', fontsize=20, loc='top')
for ax in axs:
    ax.tick_params(axis='both', which='major', labelsize=15)
plt.show()
```

### Now take the normally distributed x,y,z and sample the points in the sphere

```python
def SphericalSample(dim, radius, num_points):
    import numpy as np
    x = np.random.normal(0,1,(num_points,dim))#generate the random(gaussian) pairs
    z = np.linalg.norm(x, axis=1) #calculate the norm of the random pairs
    if (z==0).any(): #kick out any pair whose norm = 0 also remove the respectable norms
        i = np.where(z==0)
        x = np.delete(x,i)
        z = np.delete(z, i)
    #
    z = z.reshape(-1,1).repeat(x.shape[1], axis=1)# z has the following form [ [norm_0, norm_0, ..., norm_0], ... ] 

    Points = x/z * radius 
    return Points    

points1, points2, points3= [SphericalSample(3, 1,i) for i in range(500, 2000, 500)]
x1,y1,z1 = points1.T
x2,y2,z2 = points2.T
x3,y3,z3 = points3.T

display(Markdown("#### Plot the spherical distributions for number of points = (500, 1000, 1500)"))
fig = plt.figure(figsize=(27,15))
ax1 = fig.add_subplot(2,3,1, projection='3d')
ax1.scatter(x1,y1,z1, s=5)
ax1.set_xlabel('x', fontsize=20);ax1.set_ylabel('y', fontsize=20)
ax1.tick_params(axis='both', which='major', labelsize=15)
#
ax11 = fig.add_subplot(2,3,4, projection='3d')
ax11.scatter(x1,y1,z1, s=5)
ax11.set_xlabel('x', fontsize=20);ax11.set_ylabel('y', fontsize=20)
ax11.tick_params(axis='both', which='major', labelsize=15)
ax11.view_init(elev=90, azim=-90)
ax11.set_zticks([])
#
ax2 = fig.add_subplot(2,3,2, projection='3d')
ax2.scatter(x2,y2,z2, s=5)
ax2.set_xlabel('x', fontsize=20)
ax2.set_ylabel('y', fontsize=20)
ax2.tick_params(axis='both', which='major', labelsize=15)
#
ax22 = fig.add_subplot(2,3,5, projection='3d')
ax22.scatter(x2,y2,z2, s=5)
ax22.view_init(elev=90, azim=-90)
ax22.set_xlabel('x', fontsize=20)
ax22.set_ylabel('y', fontsize=20)
ax22.tick_params(axis='both', which='major', labelsize=15)
ax22.set_zticks([])

#ax2.set_axis_off()
#
ax3 = fig.add_subplot(2,3,3, projection='3d')
ax3.scatter(x3,y3,z3, s=5)
ax3.set_xlabel('x', fontsize=20)
ax3.set_ylabel('y', fontsize=20)
ax3.tick_params(axis='both', which='major', labelsize=15)
#
ax33 = fig.add_subplot(2,3,6, projection='3d')
ax33.scatter(x3,y3,z3,s=5)
ax33.view_init(elev=90, azim=-90)
ax33.set_xlabel('x', fontsize=20, loc='left')
ax33.set_ylabel('y', fontsize=20, loc='top')
ax33.tick_params(axis='both', labelsize=15)
ax33.set_zticks([])
#ax3.set_axis_off()
plt.show()

# Lets see how the numbers are distributed
display(Markdown("#### Plot the distributions of the smapled vector's components"))
#display_markdown("*Plot the distributions of the sampled vector's components*")
fig_hist, axs = plt.subplots(
    1, 3,sharey=True, tight_layout=True,  figsize=(17,7)
)

n_bins = 50
axs[0].hist(x3, bins=n_bins,histtype='step')
axs[0].set_title(r'$\frac{x}{\sqrt{x^{2}+y^{2}+z^{2}}}$', fontsize=20)
#axs[0].set_xlabel('data', fontsize=20, loc='right')
axs[0].set_ylabel('counts', fontsize=20, loc='top')
axs[1].hist(y3, bins=n_bins, histtype='step')
axs[1].set_title(r'$\frac{y}{\sqrt{x^{2}+y^{2}+z^{2}}}$', fontsize=20)
#axs[1].set_xlabel('data', fontsize=20, loc='right')
axs[2].hist(z3, bins=n_bins,  histtype='step')
axs[2].set_title(r'$\frac{z}{\sqrt{x^{2}+y^{2}+z^{2}}}$', fontsize=20)
axs[2].set_xlabel('data',fontsize=20, loc='right')
for ax in axs:
    ax.tick_params(axis='both', which='major', labelsize=15)
plt.show()
```

The distributions of each of the vector's component are flat.

## Method 2: Uniformly distribute $\theta$ and $\phi$

Now I will use a more intuitive method, well tailored for the 3dimentional case. 
By defining $\theta' = -cos{\theta}$ the elementary surface of a sphere with radius R takes can be written as 
$$dS=Rd\theta'd\phi$$. 
Thus, by sampling $\theta'$ in [-1,1] and $\phi$ in [0, 2$\pi$], $\theta=\cos^{-1}({-\theta'})$ and $\phi$ will be uniformly distributed. 
Finaly, with the values of $\theta$ and $\phi$, I will create the points 

$$
\begin{matrix}
x = R\cos{\theta}\sin{\phi}\\
y = R\sin{\theta}\sin{\phi}\\
z = R\cos{\theta}\\
\end{matrix}
$$

R, is the radius of the sphere in question

```python
pi=np.pi
```

```python
def RandAngles(num_points):
    #sample theta
    rangeTheta=(-1,1)
    theta_prime=(rangeTheta[0]-rangeTheta[-1])*np.random.random_sample(num_points,) + rangeTheta[-1]
    theta=np.arccos(-theta_prime)
    #sample phi
    rangePhi=(0, 2*pi)
    phi=(rangePhi[0]-rangePhi[-1])*np.random.random_sample(num_points,)+rangePhi[-1]
    angles = (theta, phi)
    return angles

#lets verify that the angles are distributed uniformly
theta, phi = RandAngles(1000)
fig_hist, axs = plt.subplots(
    1, 2,sharey=True, tight_layout=True,  figsize=(17,7)
)
display(Markdown("#### plot the distributions of $\\theta$ and $\phi$"))
n_bins = 50
axs[0].hist(theta, bins=n_bins,histtype='step')
axs[0].set_title(r'$\theta$ distribution', fontsize=20)
axs[1].hist(phi, bins=n_bins, histtype='step')
axs[1].set_title(r'$\phi$ distribution', fontsize = 20)
axs[0].set_ylabel('counts', fontsize=20, loc='top')
axs[1].set_xlabel('data', fontsize=20, loc='right')
for ax in axs:
    ax.tick_params(axis='both', which='major', labelsize=15)
plt.show()
```

```python
points = []
points1, points2, points3 = ([], [], [])
x1, x2, x3 = ([], [], [])
def make_cartesian(theta,phi, r):
    x = r*np.sin(theta)*np.cos(phi)
    y = r*np.sin(theta)*np.sin(phi)
    z = r*np.cos(theta)
    points=(x,y,z)
    return np.array(points)

points= [RandAngles(i) for i in range(500, 2000, 500)]
points1, points2, points3 = [make_cartesian(p[0], p[1], 1) for p in points]
#print(points1)
#x, y, z = make_cartesian(theta, phi, 1)

x1,y1,z1 = points1
x2,y2,z2 = points2
x3,y3,z3 = points3

display(Markdown("#### Plot the spherical distributions for number of points = (500, 1000, 1500)"))
fig = plt.figure(figsize=(27,15))
ax1 = fig.add_subplot(2,3,1, projection='3d')
ax1.scatter(x1,y1,z1, s=5)
ax1.set_xlabel('x', fontsize=20);ax1.set_ylabel('y', fontsize=20)
ax1.tick_params(axis='both', which='major', labelsize=15)
#
ax11 = fig.add_subplot(2,3,4, projection='3d')
ax11.scatter(x1,y1,z1, s=5)
ax11.set_xlabel('x', fontsize=20);ax11.set_ylabel('y', fontsize=20)
ax11.tick_params(axis='both', which='major', labelsize=15)
ax11.view_init(elev=90, azim=-90)
ax11.set_zticks([])
#
ax2 = fig.add_subplot(2,3,2, projection='3d')
ax2.scatter(x2,y2,z2, s=5)
ax2.set_xlabel('x', fontsize=20)
ax2.set_ylabel('y', fontsize=20)
ax2.tick_params(axis='both', which='major', labelsize=15)
#
ax22 = fig.add_subplot(2,3,5, projection='3d')
ax22.scatter(x2,y2,z2, s=5)
ax22.view_init(elev=90, azim=-90)
ax22.set_xlabel('x', fontsize=20)
ax22.set_ylabel('y', fontsize=20)
ax22.tick_params(axis='both', which='major', labelsize=15)
ax22.set_zticks([])

#ax2.set_axis_off()
#
ax3 = fig.add_subplot(2,3,3, projection='3d')
ax3.scatter(x3,y3,z3, s=5)
ax3.set_xlabel('x', fontsize=20)
ax3.set_ylabel('y', fontsize=20)
ax3.tick_params(axis='both', which='major', labelsize=15)
#
ax33 = fig.add_subplot(2,3,6, projection='3d')
ax33.scatter(x3,y3,z3, s=5)
ax33.view_init(elev=90, azim=-90)
ax33.set_xlabel('x', fontsize=20)
ax33.set_ylabel('y', fontsize=20)
ax33.tick_params(axis='both', which='major', labelsize=15)
ax33.set_zticks([])
#ax3.set_axis_off()
plt.show()

# Lets see how the numbers are distributed
display(Markdown("#### Plot the distributions of x, y, z, for R=1"))
#display_markdown("*Plot the distributions of the sampled vector's components*")
fig_hist, axs = plt.subplots(
    1, 3,sharey=True, tight_layout=True,  figsize=(17,7)
)

n_bins = 50
axs[0].hist(x3, bins=n_bins,histtype='step')
axs[0].set_title(r'$R*\sin(\theta)*\cos(\phi)$', fontsize=20)
axs[1].hist(y3, bins=n_bins, histtype='step')
axs[1].set_title(r'$R*\sin(\theta)*\sin(\phi)$', fontsize=20)
axs[2].hist(z3, bins=n_bins,  histtype='step')
axs[2].set_title(r'$R*\cos(\theta)$', fontsize=20)
axs[0].set_ylabel('counts', fontsize=20, loc='top')
axs[2].set_xlabel('data', fontsize=20, loc='right')
for ax in axs:
    ax.tick_params(axis='both', which='major', labelsize=15)
#
plt.show()
```
